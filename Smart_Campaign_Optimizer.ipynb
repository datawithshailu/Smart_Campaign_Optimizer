```python
import pandas as pd

df = pd.read_csv("campaign_data.csv")
df.head()
```

```python
# Show number of rows and columns
print("Shape of dataset:", df.shape)

# Check column names and types
print("\nColumn Info:")
print(df.info())

# Show unique values in each column (to understand categories)
print("\nUnique values per column:")
for col in df.columns:
    print(f"{col}: {df[col].unique()}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")
```

```python
# Bar chart showing campaign success by Age Group
plt.figure(figsize=(7, 5))
sns.countplot(data=df, x='Age_Group', hue='Campaign_Success', palette='Set2')
plt.title("Campaign Success by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.legend(title="Campaign Success")
plt.show()
```

```python
plt.figure(figsize=(7, 5))  # Set chart size
sns.countplot(data=df, x='Region', hue='Campaign_Success', palette='pastel')
plt.title("Campaign Success by Region")  # Chart title
plt.xlabel("Region")                     # X-axis label
plt.ylabel("Count")                      # Y-axis label
plt.legend(title="Campaign Success")     # Add legend
plt.show()                               # Show the chart
```

```python
plt.figure(figsize=(7, 5))
sns.countplot(data=df, x='Campaign_Type', hue='Campaign_Success', palette='Set3')
plt.title("Campaign Success by Campaign Type")
plt.xlabel("Campaign Type")
plt.ylabel("Count")
plt.legend(title="Campaign Success")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Product_Category', hue='Campaign_Success', palette='Set1')
plt.title("Campaign Success by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Count")
plt.legend(title="Campaign Success")
plt.show()
```
